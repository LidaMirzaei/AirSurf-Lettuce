```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load the dataset
circle_csv = pd.read_csv('/content/circles0.30.csv')

# Remove the labels from the dataset
X = circle_csv.drop('label', axis=1)

# Create a KMeans object with 2 clusters
kmeans = KMeans(n_clusters=2)

# Fit the KMeans model to the data
kmeans.fit(X)

# Get the cluster assignments for each data point
labels = kmeans.predict(X)

# Add the cluster labels back to the original dataframe
circle_csv['cluster'] = labels

# Plot the points
plt.figure(figsize=(10, 7))
plt.scatter(X['x1'], X['x2'], c=circle_csv['label'], cmap='viridis')
plt.scatter(X['x1'], X['x2'], c=circle_csv['cluster'], marker='s', cmap='viridis')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('KMeans Clustering with Original Labels')
plt.show()
```
